```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import the full file of retail distribution by quarter, broken down by gateway zip code
zip_all = pd.read_csv('retail_by_zip.csv')
```

```python
# create a dataframe that excludes the lines totaled up by zip code
no_totals = zip_all.drop(zip_all.loc[zip_all['Zip']=="TOTAL"].index)

# create dataframes summing over zip codes
state_totals_quarterly = no_totals.groupby(['State', 'Year', "Drug"]).sum()
state_total_averages = no_totals.groupby(['State', 'Drug']).mean()
state_total_averages.reset_index(inplace=True)
state_total_averages= state_total_averages.sort_values(['Drug', 'TOTAL'], ascending=False)
state_totals_quarterly.reset_index(inplace=True)
```

```python
# function to plot annual totals by zip code for a single state-drug pair
def plot_state_drug(state, drug):
    fig, ax = plt.subplots()
    labels = []
    df = no_totals[(no_totals['State']==state)&(no_totals['Drug']==drug)&(no_totals['Zip']!='TOTAL')]
    zips = df['Zip'].unique()
    for z in zips:
        labels.append(z)
        df[df['Zip']==z].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="{} Retail Distribution of {}".format(state,drug), xlabel = "Year", ylabel ="Total Grams")
    ax.legend().set_visible(False)

# function to plot the annual totals for the top X zip codes for a state-drug pair
# "top" zip codes are determined by taking the average distribution totals over the available years (2000-2016)
def plot_state_drug_top(state, drug, number):
    fig, ax = plt.subplots()
    labels = []
    df = no_totals[(no_totals['State']==state)&(no_totals['Drug']==drug)&(no_totals['Zip']!='TOTAL')]
    means = df.groupby('Zip').mean().sort_values('TOTAL', ascending=False)
    top_zips = means.head(number).index.tolist()
    for z in top_zips:
        labels.append(z)
        df[df['Zip']==z].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="{} Retail Distribution of {}, top {} zip codes".format(state, drug, number), xlabel = "Year", ylabel ="Total Grams")
    ax.legend(labels=labels, loc=2, fontsize=8)
```

```python
# function to plot the total annual distribution for a given drug by state
def plot_drug_annual(drug):
    fig, ax = plt.subplots()
    labels = []
    df = state_totals_quarterly[state_totals_quarterly['Drug']==drug]
    states = df['State'].unique()
    for s in states:
        labels.append(s)
        df[df['State']==s].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="Annual Retail Distribution of {} by State".format(drug), xlabel = "Year", ylabel ="Total Grams")
    ax.legend().set_visible(False)

# function to plot total annual distribution for a given drug by state, and label selected states
def plot_drug_annual_state_label(state_list, drug):
    fig, ax = plt.subplots()
    labels = []
    df = state_totals_quarterly[state_totals_quarterly['Drug']==drug]
    states = df['State'].unique()
    for s in states:
        if s in state_list:
            df[df['State']==s].plot(kind='line', y='TOTAL', x='Year', label=s, ax=ax)
        else:
            df[df['State']==s].plot(kind='line', y='TOTAL', x='Year', label='_nolegend_', ax=ax)
    ax.set(title="Annual Retail Distribution of {}".format(drug), xlabel = "Year", ylabel ="Total Grams")
    ax.legend(loc=2, fontsize=8)   

# function to plot only the top X states by annual distribution for a given drug
# "top" is calcuated according to the average annual totals
def plot_drug_annual_top(drug, number):
    fig, ax = plt.subplots()
    labels = []
    df = state_totals_quarterly[state_totals_quarterly['Drug']==drug]
    means = df.groupby('State').mean().sort_values('TOTAL', ascending=False)
    top_states = means.head(number).index.tolist()
    for s in top_states:
        labels.append(s)
        df[df['State']==s].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="Annual Retail Distribution of {}, Top {} States".format(drug, number), xlabel = "Year", ylabel ="Total Grams")
    ax.legend(labels=labels, loc=2, fontsize=8)
```

```python
plot_state_drug_top('WASHINGTON', 'OXYCODONE', 20)
```

```python
plot_drug_annual_state_label(state_list=['MICHIGAN'],drug="OXYCODONE")
```

```python
plot_drug_annual_top('OXYCODONE', 10)
```

```python
activity_all = pd.read_csv('activity_by_state.csv')
```

```python
activity_all.head()
```

```python
def plot_state_drug_grams_by_activity(state, drug):
    fig, ax = plt.subplots()
    labels = []
    df = activity_all[(activity_all['State']==state)&(activity_all['Drug']==drug)]
    activities = df['Business Activity'].unique()
    for a in activities:
        labels.append(a)
        df[df['Business Activity']==a].plot(kind='line', y='Total grams sold', x='Year', ax=ax)
    ax.set(title="{} Retail Distribution of {} by Business Activity".format(state,drug), xlabel = "Year", ylabel ="Total Grams Sold")
    ax.legend(labels=labels, loc=2, fontsize=4)
    
def plot_state_drug_registrants_by_activity(state, drug):
    fig, ax = plt.subplots()
    labels = []
    df = activity_all[(activity_all['State']==state)&(activity_all['Drug']==drug)]
    activities = df['Business Activity'].unique()
    for a in activities:
        labels.append(a)
        df[df['Business Activity']==a].plot(kind='line', y='Registrants', x='Year', ax=ax)
    ax.set(title="{}: Number of Retail Distribution Registrants of {}".format(state,drug), xlabel = "Year", ylabel ="Number of Registrants")
    ax.legend(labels=labels, loc=2, fontsize=8)
```

```python
plot_state_drug_grams_by_activity('FLORIDA', 'METHADONE')
```

```python
plot_state_drug_registrants_by_activity("FLORIDA", 'METHADONE')
```

```python
population_dist = pd.read_csv('distribution_by_pop.csv')
```

```python
population_dist.drop(population_dist.loc[population_dist['State']=='UNITED STATES'].index, inplace=True)
population_dist.drop(population_dist.loc[population_dist['State']=='UNITED STATES TOTAL'].index, inplace=True)
population_dist.drop(population_dist.loc[population_dist['State']=='U.S. TOTAL'].index, inplace=True)
```

```python
population_dist['State'].unique()
```

```python
# function to plot the total annual distribution for a given drug by state
def plot_drug_annual_pop(drug):
    fig, ax = plt.subplots()
    labels = []
    df = population_dist[population_dist['Drug']==drug]
    states = df['State'].unique()
    for s in states:
        labels.append(s)
        df[df['State']==s].plot(kind='line', y='TOTAL TO DATE', x='Year', ax=ax)
    ax.set(title="Annual Retail Distribution of {} by State per 100K Population".format(drug), 
           xlabel = "Year", ylabel ="Grams per 100K Population")
    ax.legend().set_visible(False)

# function to plot total annual distribution for a given drug by state, and label selected states
def plot_drug_annual_state_label_pop(state_list, drug):
    fig, ax = plt.subplots()
    labels = []
    df = population_dist[population_dist['Drug']==drug]
    states = df['State'].unique()
    for s in states:
        if s in state_list:
            df[df['State']==s].plot(kind='line', y='TOTAL TO DATE', x='Year', label=s, ax=ax)
        else:
            df[df['State']==s].plot(kind='line', y='TOTAL TO DATE', x='Year', label='_nolegend_', ax=ax)
    ax.set(title="Annual Retail Distribution of {} per 100K Population".format(drug), 
           xlabel = "Year", ylabel ="Grams per 100K Population")

    ax.legend(loc=2, fontsize=8)   

# function to plot only the top X states by annual distribution for a given drug
# "top" is calcuated according to the average annual totals
def plot_drug_annual_top_pop(drug, number):
    fig, ax = plt.subplots()
    labels = []
    df = population_dist[population_dist['Drug']==drug]
    means = df.groupby('State').mean().sort_values('TOTAL TO DATE', ascending=False)
    top_states = means.head(number).index.tolist()
    for s in top_states:
        labels.append(s)
        df[df['State']==s].plot(kind='line', y='TOTAL TO DATE', x='Year', ax=ax)
    ax.set(title="Annual Retail Distribution of {} by 100K Population, Top {} States".format(drug, number), 
           xlabel = "Year", ylabel ="Grams per 100K Population")
    ax.legend(labels=labels, loc=2, fontsize=8)
```

```python
plot_drug_annual_top_pop('FENTANYL BASE', 10)
```

```python
plot_drug_annual_pop('METHADONE')
```

```python
def plot_drug_year_pop(drug, year):
    fig, ax = plt.subplots(figsize=(10,10))
    df = population_dist[(population_dist['Drug']==drug)&(population_dist['Year']==year)].sort_values('TOTAL TO DATE')
    df.plot(kind='bar', x='State', y='TOTAL TO DATE', ax=ax)
```

```python
plot_drug_year_pop('OXYCODONE', 2011)
```
