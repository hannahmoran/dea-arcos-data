### DEA ARCOS Data - Exploratory Analysis and Visualization

The DEA publishes data annually from its Automation of Reports and Consolidated Orders System, or ARCOS. According to the DEA's website, ARCOS "monitors the flow of DEA controlled substances from their point of manufacture through commercial distribution channels to point of sale or distribution at the dispensing/retail level - hospitals, retail pharmacies, practitioners, mid-level practitioners, and teaching institutions....these transactions...are then summarized into reports which give investigators in Federal and state government agencies information which can then be used to identify the diversion of controlled substances into illicit channels of distribution. The information on drug distribution is used throughout the United States (U.S.) by U.S. Attorneys and DEA investigators to strengthen criminal cases in the courts."

So, ARCOS exists to help the government identify patterns in the manufacture and distribution of controlled substances that might indicate that these substances are being sold illegally. Annual ARCOS reports are publically available on the DEA's website, dating back to the year 2000, but unfortunately they are only available in PDF form and are dozens or even hundreds of pages long. 

I was interested in doing some data analysis and visualization on the distribution of oxycodone, an opioid painkiller that is one of the main drivers of the current prescription pain pill (and arguably heroin) addiction epidemic in the United States right now. 

The code I used to clean and process the PDF files is in another notebook if you are interested in that - it's a useful opportunity to practice data cleansing skills with pandas. 

Here I will share a few of the simple exploratory analyses I did which are the foundation of my main write-up notebook.


### Simple visualizations

To start, this data has a lot of rows and since we'll be mostly interested in seeing how distribution changes over time and comparing various geographies and business activities to each oher, I'll start with some very simple visualization functions. 

If you are new to matplotlib or want a refresher, this post was very helpful to me: http://pbpython.com/effective-matplotlib.html


```python
# first read in the three dataframes
zip_all = pd.read_csv('retail_distribution_by_zipcode.csv')
activity_all = pd.read_csv('distribution_by_activity.csv')
pop_all = pd.read_csv('distribution_by_100K_pop.csv')
```

```python
def plot_state_drug(states, drugs, by=['Total', 'Zip', 'Activity', 'Population']):
    """
    Plots the annual retail distribution of a single drug in a single state over the years in the dataset.
    Distribution is shown in grams by either gateway zip code, by business activity type, or per 100K population. 
    """
    fig, ax = plt.subplots()
    labels = []
    for state in states:
        for drug in drugs:
            if by == 'Total':
                labels.append('{} in {}'.format(drug,state))
                df = zip_all[(zip_all['State']==state) & (zip_all['Drug']==drug) & (zip_all['Zip']=="TOTAL")].sort_values('Year')
                df.plot(kind='line', y='TOTAL', x='Year', ax=ax)

                ax.set(title="Retail distribution of {} in {}, in total grams".format(', '.join(drugs), ', '.join(states)),
                      xlabel='Years', ylabel='Total grams distributed')
                ax.legend(labels=labels, loc=2, fontsize=6)
                
            elif by =='Zip':
                df = zip_all[(zip_all['State']==state) & (zip_all['Drug']==drug) & (zip_all['Zip']!="TOTAL")].sort_values('Year')
                items = df['Zip'].unique()
                for i in items:
                    labels.append(i)
                    df[df['Zip']==i].plot(kind='line', y='TOTAL', x='Year', ax=ax)

                ax.set(title="Retail distribution of {} in {}, in total grams by gateway zipcode".format(', '.join(drugs), ', '.join(states)),
                      xlabel='Years', ylabel='Total grams distributed')
                ax.legend().set_visible(False)

            elif by == 'Activity':
                df = activity_all[(activity_all['State']==state) & (activity_all['Drug']==drug)].sort_values('Year')
                items = df['Business Activity'].unique()
                for i in items:
                    labels.append(i)
                    df[df['Business Activity']==i].plot(kind='line', y='Total grams sold', x='Year', ax=ax)

                ax.set(title="Retail distribution of {} in {}, in total grams by business activity".format(', '.join(drugs), ', '.join(states)),
                      xlabel='Years', ylabel='Total grams distributed')
                ax.legend(labels=labels, loc=2, fontsize=6)

            elif by == 'Population':
                labels.append('{} in {}'.format(drug,state))
                df = pop_all[(pop_all['State']==state) & (pop_all['Drug']==drug)].sort_values('Year')

                df.plot(kind='line', y='TOTAL', x='Year', ax=ax)

                ax.set(title="Retail distribution of {} in {}, in grams per 100K population".format(', '.join(drugs), ', '.join(states)),
                      xlabel='Years', ylabel='Grams distributed per 100K population')
                ax.legend().set_visible(False)
                ax.legend(labels=labels, loc=2, fontsize=6)

            else:
                print("Invalid selection - please indicate if the data should be shown by zip code, activity, or population")

        
        
```

```python
plot_state_drug(by='Total', drugs=['OXYCODONE'], states=["FLORIDA"])
plot_state_drug(by='Population', drugs=['OXYCODONE'], states=['FLORIDA']),
plot_state_drug(by='Activity', drugs=['OXYCODONE'], states=['FLORIDA']),
plot_state_drug(by='Zip', drugs=['OXYCODONE'], states=['FLORIDA'])
```

If you know anything about the prescription pain pill epidemic, you might know that Florida was, for a handful of years, a hotbed of what are known as "pill mills." These are clinics where patients can expect to receive large prescriptions, on a regular basis, of opioid painkillers, without having to provide much evidence of real medical need. Florida became a destination for addicts and drug dealers because of the easy access to these prescription medications. This really kicked off around 2008-2009 and was sharply curtailed in 2011 by Florida's introduction of a prescription drug monitoring database and tougher restrictions on the pill mill clinics. 

All of this is clearly visible in the charts above. The zip code chart is a little heavy, given how many gateway zip codes we have in the data. It would be nice to be able to see the top zip codes by distribution amount, and put some labels on them as well.

```python
def plot_top_zips(state, drug, number):
    fig, ax = plt.subplots()
    labels = []
    df = zip_all[(zip_all['State']==state)&(zip_all['Drug']==drug)&(zip_all['Zip']!='TOTAL')].sort_values('Year')
    means = df.groupby('Zip').mean().sort_values('TOTAL', ascending=False)
    top_zips = means.head(number).index.tolist()
    for z in top_zips:
        labels.append(z)
        df[df['Zip']==z].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="Retail Distribution of {} in {}, in total grams for the top {} zip codes".format(drug, state, number), 
           xlabel = "Year", ylabel ="Total Grams")
    ax.legend(labels=labels, loc=2, fontsize=8)
    
```

```python
plot_top_zips('FLORIDA', 'OXYCODONE', 5)
```

Now we can see that two of the zip codes groups with the highest levels of oxycodone distribution during that peak in 2010 are 334 and 336. Googling these gateway zip codes will show you that they include Palm Beach, Hillsborough, and Broward counties, where the cities of West Palm Beach, Tampa, Fort Lauderdale, Hallandale, and Hollywood are located. Further research will show that these areas were indeed hotbeds of pill mill activity - John Temple's book "American Pain" tells the unbelievable story of the two men who ran some of the largest pill mills in South Florida, including clinics in West Palm Beach. 

Finally, I might like to be able to see distribution activity in specific zip codes.

```python
def plot_zips(state, drug, zips):
    fig, ax = plt.subplots()
    labels = []
    df = zip_all[(zip_all['State']==state)&(zip_all['Drug']==drug)&(zip_all['Zip']!='TOTAL')].sort_values('Year')
    for z in zips:
        labels.append(z)
        df[df['Zip']==z].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="Retail Distribution of {} in {}, in total grams for selected zip codes".format(drug, state), 
           xlabel = "Year", ylabel ="Total Grams")
    ax.legend(labels=labels, loc=2, fontsize=8)
```

Out of curiosity, I'd like to see what the oxycodone activity looks like in the zip code that houses my hometown in Michigan.

```python
plot_zips("MICHIGAN", 'OXYCODONE', ['481'])
```

Methadone and buprenorphine are two drugs that are used to treat opioid addiction. They are opioids themselves, and can also be used to treat pain, but they do not produce the same highs that other opioids like oxycodone and heroin do. For an addict, they prevent withdrawal symptoms, including reducing opioid cravings and even blocking the euphoric effects that other opioid drugs might have produced for that person. They are also metabolized extremely slowly as compared to other opioids, and are generally prescribed to be taken once a day for this type of maintenance therapy. Many people, especially those who have struggled to remain clean after going through detox and rehab programs, successfully use them as a longterm form of treatment for their addiction.

Although these medications are controversial, we might expect to see a rise in their use in areas where opioid distribution is extremely high (given that we assume very high rates of opioid distribution are also corresponding to rising rates of abuse and addiction). 

```python
plot_state_drug(states=["FLORIDA"], drugs=["METHADONE", 'BUPRENORPHINE'], by='Total')
plot_state_drug(states=["FLORIDA"], drugs=["METHADONE", 'BUPRENORPHINE', 'OXYCODONE'], by='Total')
```

Oddly, the peak of methadone distribution in Florida doesn't seem to correspond as one might expect with the peak of oxycodone, although distribution of buprenorphine has risen steadily over the years for which we have data. Note how little of these drugs is being distributed in the state as compared to oxycodone.

There's a much more striking finding when we look by zip code for methadone - one single gateway zip code, 321, was driving a bit less than half the oxycodone distribution of the entire state between 2006 and 2011. I spent quite a while trying to understand what could be driving this. The last chart shows that most of the demand was coming from treatment programs and pharmacies.

Gateway zip code 321 includes the city of Daytona Beach. A news story in a local newspaper questioning the profit motives of the area's methadone clinics seems to indicate that there was some amount of concern with diversion of methadone being prescribed in these clinics. Diversion means that the drug is being obtained in a legitimate (or at least legal) manner but is then being sold on the streets to people who are likely abusing it. Methadone and buprenorphine are usually tightly controlled, and those who are on a maintenance program have to visit a clinic each day to get their dose and must take it under supervision. This is both for the safety of the patient (due to the length of time for which they remain in the body, there is a greater danger of accidental overdose due to buildup of the drug if not taken properly) and to ensure that the medication isn't being diverted. However, some clinics will allow patients "takeaways," or take home doses, usually after they have been a patient for a certain period and have demonstrated a commitment to responsibly taking their medication. It's certainly possible that a less scrupulous clinic might not be so watchful. 

One of the only other specific references to methadone in the communities in 321 was a horrible story about a young teenager overdosing on discarded methadone found in the dumpster of a pharmaceutical distributor. Unfortunately I could find very little additional information about the distributor or the victim - I found a brief reference in a local newspaper suggesting that the boy's family sued the company, but nothing else. 

In short, I did not find much information and certainly nothing that would give a satisfactory explanation for why there was so much methadone being moved in this one small area of Florida during these years. Perhaps someone with more skills in investigative journalism can find out the answer.
