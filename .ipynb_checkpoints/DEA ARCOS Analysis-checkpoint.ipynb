```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import the full file of retail distribution by quarter, broken down by gateway zip code
zip_all = pd.read_csv('retail_by_zip.csv')

# read in the full file of distribution by business activity annually for each state
activity_all = pd.read_csv('activity_by_state.csv')

# read in some basic annual data on drug poisoning mortality by state
# from https://catalog.data.gov/dataset/drug-poisoning-mortality-u-s-and-state-trends-united-states-20022014/resource/57f97565-b371-422c-8466-800d4f26f7d9
drug_deaths_summary = pd.read_csv('drug_deaths.csv')
```

```python
# create a dataframe that excludes the lines totaled up by zip code
no_totals = zip_all.drop(zip_all.loc[zip_all['Zip']=="TOTAL"].index)

# create dataframes summing over zip codes
state_totals_quarterly = no_totals.groupby(['State', 'Year', "Drug"]).sum()
state_total_averages = no_totals.groupby(['State', 'Drug']).mean()
state_total_averages.reset_index(inplace=True)
state_total_averages= state_total_averages.sort_values(['Drug', 'TOTAL'], ascending=False)
state_totals_quarterly.reset_index(inplace=True)
```

```python
# function to plot age-adjusted death rate, per 100,000 US std population 2000
def plot_age_adj_od_deaths(state_list):
    fig, ax = plt.subplots()
    df = drug_deaths_summary
    for s in state_list:
        df[df['State']==s].plot(kind='line', y="Age-adjusted Rate", x='Year', ax=ax, label=s)
    df[df['State']==state_list[0]].plot(kind='line', y='US Age-adjusted Rate', x='Year', ax=ax, label = "US Death Rate")
    ax.set(title='Drug Overdose Deaths in {}'.format(state_list), xlabel='Year', ylabel='Age-Adjusted Death Rate (per 100,000)')
    ax.legend(loc=2, fontsize=8)

def plot_age_adj_od_deaths_top(number):
    fig, ax = plt.subplots()
    labels = []
    df = drug_deaths_summary
    means = df.groupby('State').mean().sort_values('Age-adjusted Rate', ascending=False)
    top_states = means.head(number).index.tolist()
    for s in top_states:
        labels.append(s)
        df[df['State']==s].plot(kind='line', y='Age-adjusted Rate', x='Year', ax=ax)
    labels.append('US Death Rate')
    df[df['State']==top_states[0]].plot(kind='line', y='US Age-adjusted Rate', x='Year', ax=ax)
    ax.set(title='Top {} States by Drug Overdose Deaths'.format(number), xlabel='Year', ylabel='Age-Adjusted Death Rate (per 100,000)')
    ax.legend(labels=labels, loc=2, fontsize=8)
```

```python
# function to plot annual totals by zip code for a single state-drug pair
def plot_state_drug(state, drug):
    fig, ax = plt.subplots()
    labels = []
    df = no_totals[(no_totals['State']==state)&(no_totals['Drug']==drug)&(no_totals['Zip']!='TOTAL')]
    zips = df['Zip'].unique()
    for z in zips:
        labels.append(z)
        df[df['Zip']==z].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="{} Retail Distribution of {}".format(state,drug), xlabel = "Year", ylabel ="Total Grams")
    ax.legend().set_visible(False)

# function to plot the annual totals for the top X zip codes for a state-drug pair
# "top" zip codes are determined by taking the average distribution totals over the available years (2000-2016)
def plot_state_drug_top(state, drug, number):
    fig, ax = plt.subplots()
    labels = []
    df = no_totals[(no_totals['State']==state)&(no_totals['Drug']==drug)&(no_totals['Zip']!='TOTAL')]
    means = df.groupby('Zip').mean().sort_values('TOTAL', ascending=False)
    top_zips = means.head(number).index.tolist()
    for z in top_zips:
        labels.append(z)
        df[df['Zip']==z].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="{} Retail Distribution of {}, top {} zip codes".format(state, drug, number), xlabel = "Year", ylabel ="Total Grams")
    ax.legend(labels=labels, loc=2, fontsize=8)
```

```python
# function to plot annual distribution of a drug in a given state, broken down by business activity
def plot_state_drug_grams_by_activity(state, drug):
    fig, ax = plt.subplots()
    labels = []
    df = activity_all[(activity_all['State']==state)&(activity_all['Drug']==drug)]
    activities = df['Business Activity'].unique()
    for a in activities:
        labels.append(a)
        df[df['Business Activity']==a].plot(kind='line', y='Total grams sold', x='Year', ax=ax)
    ax.set(title="{} Retail Distribution of {} by Business Activity".format(state,drug), xlabel = "Year", ylabel ="Total Grams Sold")
    ax.legend(labels=labels, loc=2, fontsize=8)

# function to plot number of registrants registered to distribute a given drug each year in a given state    
def plot_state_drug_registrants_by_activity(state, drug):
    fig, ax = plt.subplots()
    labels = []
    df = activity_all[(activity_all['State']==state)&(activity_all['Drug']==drug)]
    activities = df['Business Activity'].unique()
    for a in activities:
        labels.append(a)
        df[df['Business Activity']==a].plot(kind='line', y='Registrants', x='Year', ax=ax)
    ax.set(title="{}: Number of Retail Distribution Registrants of {}".format(state,drug), xlabel = "Year", ylabel ="Number of Registrants")
    ax.legend(labels=labels, loc=2, fontsize=8)

# function to plot average grams of a drug distributed annually in a state, broken down by type of registrant
def plot_state_drug_avg_per_registrant(state, drug):
    fig, ax = plt.subplots()
    labels = []
    df = activity_all[(activity_all['State']==state)&(activity_all['Drug']==drug)]
    activities = df['Business Activity'].unique()
    for a in activities:
        labels.append(a)
        df[df['Business Activity']==a].plot(kind='line', y='Avg grams/registrant', x='Year', ax=ax)
    ax.set(title="{}: Average grams of {} distributed per registrant".format(state,drug), xlabel = "Year", ylabel ="Average Grams per Resident")
    ax.legend(labels=labels, loc=2, fontsize=8)
```

```python
# function to plot the total annual distribution for a given drug by state
def plot_drug_annual(drug):
    fig, ax = plt.subplots()
    labels = []
    df = state_totals_quarterly[state_totals_quarterly['Drug']==drug]
    states = df['State'].unique()
    for s in states:
        labels.append(s)
        df[df['State']==s].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="Annual Retail Distribution of {} by State".format(drug), xlabel = "Year", ylabel ="Total Grams")
    ax.legend().set_visible(False)

# function to plot total annual distribution for a given drug by state, and label selected states
def plot_drug_annual_state_label(state_list, drug):
    fig, ax = plt.subplots()
    labels = []
    df = state_totals_quarterly[state_totals_quarterly['Drug']==drug]
    states = df['State'].unique()
    for s in states:
        if s in state_list:
            df[df['State']==s].plot(kind='line', y='TOTAL', x='Year', label=s, ax=ax)
        else:
            df[df['State']==s].plot(kind='line', y='TOTAL', x='Year', label='_nolegend_', ax=ax)
    ax.set(title="Annual Retail Distribution of {}".format(drug), xlabel = "Year", ylabel ="Total Grams")
    ax.legend(loc=2, fontsize=8)   

# function to plot only the top X states by annual distribution for a given drug
# "top" is calcuated according to the average annual totals
def plot_drug_annual_top(drug, number):
    fig, ax = plt.subplots()
    labels = []
    df = state_totals_quarterly[state_totals_quarterly['Drug']==drug]
    means = df.groupby('State').mean().sort_values('TOTAL', ascending=False)
    top_states = means.head(number).index.tolist()
    for s in top_states:
        labels.append(s)
        df[df['State']==s].plot(kind='line', y='TOTAL', x='Year', ax=ax)
    ax.set(title="Annual Retail Distribution of {}, Top {} States".format(drug, number), xlabel = "Year", ylabel ="Total Grams")
    ax.legend(labels=labels, loc=2, fontsize=8)
```

John Temple's 2015 book "American Pain" tells the story of two young Floridians who became millionaires running what have become known as "pill mills:" clinics whose purpose is to dispense huge amounts of prescription medication (usually painkillers) as quickly and with as few questions asked as possible. Between 2007 and 2010 when their clinics were active, Florida doctors were dispensing as much as fourteen times as much oxycodone as doctors in the rest of the US combined. <br><br>
Purdue Pharma introduced OxyContin in 1996 and began an aggressive and wildly successful marketing campaign. Between 1996 and 2000, sales grew from \$48 million to nearly \$1.1 billion. 

OxyContin is Purdue Pharma's branded preparation of oxycodone, which is a controlled substance. Under the Controlled Substances Act of 1970, manufacturers and distributors of controlled substances are required to report transactions involving these substances to the Attorney General. This occurs via the DEA's Automation of Reports and Consolidated Orders System, or ARCOS. According to the DEA's website, ARCOS "monitors the flow of DEA controlled substances from their point of manufacture through commercial distribution channels to point of sale or distribution at the dispensing/retail level - hospitals, retail pharmacies, practitioners, mid-level practitioners, and teaching institutions....these transactions...are then summarized into reports which give investigators in Federal and state government agencies information which can then be used to identify the diversion of controlled substances into illicit channels of distribution. The information on drug distribution is used throughout the United States (U.S.) by U.S. Attorneys and DEA investigators to strengthen criminal cases in the courts."

So, ARCOS exists to help the government identify patterns in the manufacture and distribution of controlled substances that might indicate that these substances are being sold illegally. Annual ARCOS reports are publically available on the DEA's website in PDF format, dating back to the year 2000. For each year, I've digitized Report 1, which details controlled substance quarterly distribution in grams by US zip code prefix (the first 3 numbers of the zip code), and Report 5, which breaks annual distribution in grams down by state and by business activity (hospital, pharmacy, etc.). You can find csv copies of these on Kaggle and on my Github profile. 

So what has oxycodone distribution looked like in Florida over the past 16 years?

```python
plot_state_drug_top('FLORIDA', 'OXYCODONE', 20)
```

Zip code prefixes 334 and 336 correspond Palm Beach and Hillsborough counties, where the cities of West Palm Beach and Tampa are located, respectively. 330 and 333 zip codes are located in Broward County, where you'll find South Florida cities including Fort Lauderdale, Hallandale Beach, and Hollywood. The pill mills storied in "American Pain" were located in Palm Beach and Broward counties. 

It's easy to see the massive spike in oxycodone distribution, peaking in 2010 and dropping off sharply thereafter as Florida started to regulate pain clinics, with further decline after 2011 when the state introduced its prescription drug monitoring program in 2011. 

But the true magnitude of distribution isn't clear until you see Florida's activity against other states. 

```python
plot_drug_annual_state_label(state_list=['FLORIDA'],drug="OXYCODONE")
```

```python
plot_drug_annual_top('OXYCODONE', 5)
```

For reference...in 2010, California's population was more than double that of Florida (37.25 million to 18.84 million).  

Florida's population could not possibly have been absorbing all of the oxycodone being prescribed, and in fact, it was not. An influx of addicts and dealers from states including West Virginia, Kentucky, and Tennessee (among others) began traveling to Florida specifically for the easy access to opiate painkillers. This is a well-documented phenomenon. "American Pain" and "Dreamland" tell the stories of individuals who made this journey on a near-weekly basis. Some flew and some drove - direct flights between Huntington, WV and South Florida became known as the "OxyContin Express," a nickname also applied to the stretch of I-75 connecting the states. 

Given the overwhelming volume of prescription pain pills being prescribed in Florida, how have overdose deaths in the state compared to the national death rate? 

```python
plot_age_adj_od_deaths('Florida')
```

Florida's overdose death rate is higher than the US overall rate, but not wildly so, and even dips below the national rate briefly in the years following 2011, when the prescription drug monitoring program was introduced. Keep in mind these figures are not limited to overdose deaths caused by opiates.

```python
plot_age_adj_od_deaths_top(10)
```

In fact, Florida doesn't even crack the top 10 (highest average annual drug overdose death rate between 2000-2015)  - but West Virginia leads, with Kentucky and Tennessee in the 5th and 10th spots, respectively. 

```python
plot_age_adj_od_deaths(['Florida', 'West Virginia', 'Kentucky', 'Tennessee'])
```
