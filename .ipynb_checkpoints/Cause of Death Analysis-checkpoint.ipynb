```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cod = pd.read_csv('cod_aggregate.csv')
```

```python
cod_age = pd.read_csv('cod_age.csv')
cod_age.drop(['Unnamed: 0'],axis=1, inplace=True)
cod_urban = pd.read_csv('cod_urban.csv')
cod_gender = pd.read_csv('cod_gender.csv')
cod_race = pd.read_csv('cod_race.csv')
cod_hispanic = pd.read_csv('cod_hispanic.csv')
```

```python
causes = ['Cancer', 'Heart Diseases', 'Diabetes', 'Heroin Overdose', 'Other Opioid Overdose', 'Motor Vehicle Accident', 'Drug Overdose']
```

```python
def plot_state_gender_by_cod(state, gender):
    
    if gender=='Female':
        g='Women'
    else:
        g="Men"
    fig, ax = plt.subplots()
    df = cod_gender[cod_gender['State']==state]
    labels = []
    for cause in causes:
        df[df['Gender']==gender].plot(kind='line', x='Year', y='{} Age Adjusted Rate'.format(cause), ax=ax)
        labels.append(cause)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = "Age-Adjusted Death Rate per 100K Population", xlabel="Year", 
           title='{} Deaths Among {}'.format(state, g))

def plot_state_cod_by_gender(state, cod):
    fig, ax = plt.subplots()
    df = cod_gender[cod_gender['State']==state]
    labels = []
    for g in ['Male', 'Female']:
        df[df['Gender']==g].plot(kind='line', x="Year", y='{} Age Adjusted Rate'.format(cod), ax=ax)
        labels.append(g)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = 'Age Adjusted Death Rate per 100K Population', xlabel="Year", 
           title='{} Deaths by Gender in {}'.format(cod, state))
```

```python
def plot_state_race_by_cod(state, race):

    fig, ax = plt.subplots()
    df = cod_race[cod_race['State']==state]
    labels = []
    for cause in causes:
        df[df['Race']==race].plot(kind='line', x='Year', y='{} Age Adjusted Rate'.format(cause), ax=ax)
        labels.append(cause)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = "Age-Adjusted Death Rate per 100K Population", xlabel="Year", 
           title='{} Deaths Among {}s'.format(state, race))

def plot_state_cod_by_race(state, cod):
    fig, ax = plt.subplots()
    df = cod_race[cod_race['State']==state]
    labels = []
    for r in df['Race'].unique():
        df[df['Race']==r].plot(kind='line', x="Year", y='{} Deaths'.format(cod), ax=ax)
        labels.append(r)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = 'Deaths', xlabel="Year", 
           title='{} Deaths by Race in {}'.format(cod, state))
```

```python
def plot_state_group_by_cod(state, age_group):
    fig, ax = plt.subplots()
    df = cod_age[cod_age['State']==state]
    labels = []
    for cause in causes:
        df[df['Age Group']==age_group].plot(kind='line', x='Year', y='{} Deaths'.format(cause), ax=ax)
        labels.append(cause)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = "Deaths", xlabel="Year", title='{} Deaths Among {}-olds'.format(state, age_group[0:-1]))
        
```

```python
def plot_state_urb_by_cod(state, urb):

    fig, ax = plt.subplots()
    df = cod_urban[cod_urban['State']==state]
    labels = []
    for cause in causes:
        df[df['2013 Urbanization']==urb].plot(kind='line', x='Year', y='{} Deaths'.format(cause), ax=ax)
        labels.append(cause)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = "Deaths", xlabel="Year", 
           title='{} Deaths in {} Areas'.format(state, urb))

def plot_state_cod_by_urb(state, cod):
    fig, ax = plt.subplots()
    df = cod_urban[cod_urban['State']==state]
    labels = []
    for u in df['2013 Urbanization'].unique():
        df[df['2013 Urbanization']==u].plot(kind='line', x="Year", y='{} Deaths'.format(cod), ax=ax)
        labels.append(u)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = 'Deaths', xlabel="Year", 
           title='{} Deaths by Race in {}'.format(cod, state))
```

```python
def plot_state_hisp_by_cod(state, hisp):

    fig, ax = plt.subplots()
    df = cod_hispanic[cod_hispanic['State']==state]
    labels = []
    for cause in causes:
        df[df['Hispanic Origin']==hisp].plot(kind='line', x='Year', y='{} Age Adjusted Rate'.format(cause), ax=ax)
        labels.append(cause)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = "Age-Adjusted Death Rate per 100K Population", xlabel="Year", 
           title='{} Deaths Among {} Population'.format(state, hisp))

def plot_state_cod_by_hisp(state, cod):
    fig, ax = plt.subplots()
    df = cod_hispanic[cod_hispanic['State']==state]
    labels = []
    for h in df['Hispanic Origin'].unique():
        df[df['Hispanic Origin']==h].plot(kind='line', x="Year", y='{} Deaths'.format(cod), ax=ax)
        labels.append(h)
    ax.legend(labels=labels, loc=2, fontsize=8)
    ax.set(ylabel = 'Deaths', xlabel="Year", 
           title='{} Deaths by Hispanic Origin in {}'.format(cod, state))
```

```python
plot_state_group_by_cod('West Virginia', '35-44 years')
```

```python
plot_state_group_by_cod('Kentucky', '25-34 years')
```

```python
plot_state_gender_by_cod(gender='Male', state="West Virginia")
```

```python
plot_state_cod_by_gender(state="Kentucky", cod='Other Opioid Overdose')
```

```python
plot_state_race_by_cod(race='Black or African American', state="Florida")
```

```python
plot_state_cod_by_race(cod='Drug Overdose', state="Alabama")
```

```python
plot_state_urb_by_cod(state="West Virginia", urb='Small Metro')
```

```python
plot_state_cod_by_urb(cod="Drug Overdose", state="Pennsylvania")
```

```python
plot_state_cod_by_hisp(cod='Heroin Overdose', state="Florida")
```

```python
plot_state_hisp_by_cod(hisp='Hispanic or Latino', state='Florida')
```
